```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.read_csv('marketing_campaign.csv', sep=';')
dataset
```

```python
dataset.info()
```

Dataset ini merupakan data responden terhadap campaign yang diberikan. Tedapat missing value pada data Income, dan untuk selanjutnya dilakukan prosesing terhadap data yang kosong.

```python
dataset.describe().T
```

Dari hasil statistik terlihat nilai mean dan median pada fitur income tidak terlihat terlalu signifikan perbedaan. Untuk missing value akan diisi oleh data median

```python
df = dataset.copy()
df
```

```python
df['Income'] = df.Income.fillna(df.Income.median())
```

```python
df.info()
```

Data df menunjukkan bahwa fitur income telah dilakukan prosesing median


```python
df.columns
```

# EDA

```python
from matplotlib import rcParams

rcParams['figure.figsize'] = 12, 10
rcParams['lines.linewidth'] = 3
rcParams['xtick.labelsize'] = 'x-large'
rcParams['ytick.labelsize'] = 'x-large'
```

```python
numerical = df[['Year_Birth','Income','Recency']]
numerical=numerical.columns
numerical
```

```python
products = df[['MntWines', 'MntFruits',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds']]
products=products.columns
products
```

```python
purchase=df[['NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']]
purchase=purchase.columns
purchase
```

```python
import warnings
warnings.filterwarnings('ignore')
for i in range (0,len(numerical)):
    plt.subplot(1,len(numerical),i+1)
    sns.boxplot(y=df[numerical[i]], color='green', orient='v')
    plt.tight_layout() 
```

```python
import warnings
warnings.filterwarnings('ignore')
for i in range (0,len(numerical)):
    plt.subplot(2,2,i+1)
    sns.histplot(x=df[numerical[i]], color='green')
    plt.tight_layout() 
```

```python
import warnings
warnings.filterwarnings('ignore')
for i in range (0,len(numerical)):
    plt.subplot(2,2,i+1)
    sns.kdeplot(x=df[numerical[i]], color='green')
    plt.tight_layout() 
```

Fitur Year_Birth merupakan fitur dengan negative distribution, Income positve distribution, dan Recency normal distribution

```python
import warnings
warnings.filterwarnings('ignore')
for i in range (0,len(products)):
    plt.subplot(1,len(products),i+1)
    sns.boxplot(y=df[products[i]], color='blue', orient='v')
    plt.tight_layout() 
```

```python
import warnings
warnings.filterwarnings('ignore')
for i in range (0,len(products)):
    plt.subplot(2,3,i+1)
    sns.histplot(x=df[products[i]], color='blue')
    plt.tight_layout() 
```

```python
import warnings
warnings.filterwarnings('ignore')
for i in range (0,len(products)):
    plt.subplot(2,3,i+1)
    sns.kdeplot(x=df[products[i]], color='blue')
    plt.tight_layout() 
```

Fitur products seluruhnya terdistribusi positif dan memiliki banyak pencilan. 

```python
import warnings
warnings.filterwarnings('ignore')
for i in range (0,len(purchase)):
    plt.subplot(1,len(purchase),i+1)
    sns.boxplot(y=df[purchase[i]], color='red', orient='v')
    plt.tight_layout() 
```

```python
categorial1=df[['Education','Marital_Status','Kidhome', 'Teenhome']]
categorial1 = categorial1.columns
categorial1
```

```python
categorial2 = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4',
       'AcceptedCmp5', 'Complain', 'Response']]
categorial2 = categorial2.columns
categorial2
```

```python
for i in range (0, len(categorial1)):
    plt.subplot(1,len(categorial1),i+1)
    sns.countplot(x=df[categorial1[i]], hue=df['Response'], orient='h')
    plt.xticks(rotation=90)
    plt.tight_layout()
```

```python
for i in range (0, len(categorial2)):
    plt.subplot(1,len(categorial2),i+1)
    sns.countplot(x=df[categorial2[i]], hue=df['Response'], orient='h')
    plt.xticks(rotation=90)
    plt.tight_layout()
```

Terdapat data yang imbalance pada customer yang menerima dan menolak campaign.

```python
dfa = df[['Year_Birth','Income','MntWines', 'MntFruits',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']]
dfa
```

```python
sns.heatmap(dfa.corr(), cmap='Blues', annot=True, fmt='.2f')
plt.show()
```

fitur numcatalogpurchase dan mntmeatproduct merupakan fitur dengan korelasi yang tinggi. NumCatalogPurchase artinya jumlah pembelian melalui catalog, mntmeatproducts adalah jumlah unit produk daging yang dibeli.

```python
sns.pairplot(dfa, diag_kind='kde')
plt.show()
```
